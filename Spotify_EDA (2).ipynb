# **Load Libraries**



* Below are the libraries that are used in order to perform EDA (Exploratory data analysis)





```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#**Load Data**

*   Loading the data into the pandas data frame is certainly one of the most important steps in EDA, as we can see that the value from the data set is comma-separated. So all we have to do is to just read the CSV into a data frame and pandas data frame does the job for us.
*   print first 10 rows of the dataframe.

```python
df = pd.read_csv('Top-50-musicality-global (1).csv')
df.head(10)

```

 * print last 10 rows of the dataframe.

```python
df.tail(10)
```

#**Columns Names of the Data**

```python
df.columns
```

#**Checking the types of data**

* The info() method provides a compact summary of the DataFrame's structure and content.
* It includes information about the data types of each column, the number of non-null (non-missing) values, and memory usage.

```python
df.info()
```

#**Dropping irrelevant columns**

* The process of dropping irrelevant columns involves removing such columns from the dataset to streamline analysis and enhance interpretability.
* This step is crucial for reducing dimensionality, improving computational efficiency, and focusing attention on the most informative features.

```python
df=df.drop(['Key','Unnamed: 0','Markets','TSignature','Country','Mode'],axis=1)
df
```

# **Checking null values**

* Null values, also known as missing values, can significantly impact the quality and reliability of data analysis and modeling. Identifying and handling null values is a fundamental part of EDA.

```python
df.isnull().sum()
```

#**Remove null values**

* Missing values can potentially interfere with statistical analyses and machine learning models, so addressing them through removal is a common practice during the EDA phase.

```python
df.dropna(inplace=True)
```

#**Add relevant columns**

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

#**Checking for duplicates**

* The duplicated() method is applied to a DataFrame to identify duplicate rows, and when followed by .sum(), it counts the total number of duplicated rows in the dataset.

```python
df.duplicated().sum()
```

#**Remove Duplicates**

```python
df=df.drop_duplicates()
df.shape
```

#**Checking for unique values**

* It is particularly useful in EDA to gain insights into the cardinality of categorical features and assess the diversity of values in numerical columns.

```python
df.nunique()
```

#**Set duration milliseconds to seconds**

```python
df['duration_s']=df['duration'].apply(lambda x:round(x/1000))
df.drop('duration',inplace=True,axis=1)
df.head()
```

#**Renaming the columns**

* It is beneficial for enhancing the readability of the dataset, aligning column names with analytical objectives, or ensuring consistency in naming conventions.

```python
df=df.rename(columns={'duration':'Duration'})
```

#**Visualization for Data Analysis**

* Scatterplot with x axis Enargy and y axis with Loudness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df,y='Loudness',x='Energy',color='c').set(title='Loudness vs Energy correlation')
```

* Pie chart

```python
sns.set(rc={"figure.figsize":(6,6)})
plt.pie(df["year"].value_counts(),labels=df["year"].unique(),autopct="%.2f",shadow=True)
plt.show()
```

* Bar graph


```python
plt.bar(df['Popularity'],df['Danceability'], color='skyblue')
plt.xlabel('Popularity')
plt.ylabel('Danceability')
plt.title('Popularity vs Danceability Graph')
plt.show()
```

* Pairplot

```python
dc=df.loc[:,['Popularity','Danceability','Acousticness','Energy','Liveness','Loudness']]
sns.set(rc={"figure.figsize":(12,10)})
sns.pairplot(dc)
```

* Histplot



```python
sns.set(rc={"figure.figsize":(7,7)})
sns.histplot(x="Popularity",data=df,kde=True)
plt.show()
```

* Line plot

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year']=df['Date'].dt.year
sns.lineplot(x="year",y="duration_s",data=df)
plt.title('year vs duration')
```

* Correlation Heatmap

```python
corr_df=df.corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,cmap='inferno',linewidths=0.5,linecolor="white")
heatmap.set_title("correlation between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

* Scatterplot with x axis Acousticness and y axis with Popularity

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df,y='Popularity',x='Acousticness',color='b').set(title='Popularity vs Acousticness correlation')
```

* Box plot

* Before outlares boxplot is given below

```python
sns.boxplot(x=df['year'],y=df['Energy'],data=df,palette='husl')
plt.show()
```

* After outlares boxplot is given below

```python
v1=df.quantile(0.25)
v2=df.quantile(0.75)
v3=v2-v1
print(v3)
```

* Convert datetime to object

```python
df['Date'] = df['Date'].astype('str')
df.head(10)
```

```python
df=df[~((df<(v1-1.5*v3))|(df>(v2+1.5*v3))).any(axis=1)]
df.shape
```

```python
sns.boxplot(x=df['year'],y=df['Energy'],data=df,palette='husl')
plt.show()
```
